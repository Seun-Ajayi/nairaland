```python
result = collection.find(
  {},
  { 'topic_id': 1, 'topic': 1, 'url': 1, 'comments': { 'text': 1, 'is_quote': 1, 'pageId': 1, 'quoted_text': 1, 'sex': 1 }, '_id': 0 }
).limit(10)
```

```python
for document in result:
    pp.pprint(document)
    
```

```python
# Assuming `result` is your list of documents
result_list = []

for document in result:
    # Extract relevant information from the document
    document_data = {
        'topic_id': document.get('topic_id', ''),
        'topic': document.get('topic', ''),
        'url': document.get('url', ''),
        'comments': []
    }

    # Extract information from each comment in the comments array
    for comment in document.get('comments', []):
        comment_data = {
            'text': comment.get('text', ''),
            'is_quote': comment.get('is_quote', False),
            'pageId': comment.get('pageId', 0),
            'quoted_text': comment.get('quoted_text', ''),
            'sex': comment.get('sex', '')
        }

        # Append the comment data to the comments list
        document_data['comments'].append(comment_data)

    # Append the document data to the result list
    result_list.append(document_data)
```

```python

query = query = [
    {
        "$match": {
            "comments.is_quote": True,
            "comments.quoted_text": {"$ne": ""}
        }
    },
    {
        "$project": {
            "topic_id": 1,
            "topic": 1,
            "url": 1,
            "comments": {
                "$filter": {
                    "input": "$comments",
                    "cond": {"$eq": ["$$this.is_quote", True]}
                }
            }
        }
    },
    {
        "$limit": 50
    }
]

result = collection.aggregate(query)

```

```python
for document in result:
    pp.pprint(document)
    
```

```python
# Assuming `result` is your list of documents
result_list = []

for document in result:
    # Extract relevant information from the document
    document_data = {
        'topic_id': document.get('topic_id', ''),
        'topic': document.get('topic', ''),
        'url': document.get('url', ''),
        'comments': []
    }

    # Extract information from each comment in the comments array
    for comment in document.get('comments', []):
        comment_data = {
            'text': comment.get('text', ''),
            'is_quote': comment.get('is_quote', False),
            'quoted_text': comment.get('quoted_text', ''),
        }

        # Append the comment data to the comments list
        document_data['comments'].append(comment_data)

    # Append the document data to the result list
    result_list.append(document_data)
```
